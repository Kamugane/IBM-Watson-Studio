###  Sample Math Operation...Conversion of Degree Celsius unit to Fahrenheit unit

```python
# Temperature in celsius degree
celsius = 30
 
# Converting the temperature to fehrenheit using the formula
fahrenheit = (celsius * 1.8) + 32
 
# printing the result
print('%.2f Celsius is equivalent to : %.2f Fahrenheit' %(celsius, fahrenheit))
```
